```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
```

```python
data = pd.read_csv('../dEtaOptimization/dEtaOptimization.csv', skiprows=2)
data['Value'] = data['Value'] / 10
```

```python
fig = plt.figure(figsize=(7, 5))
plt.plot(data['Value'], data['Significance'], 'o-')
plt.title(u'Significance vs $\Delta \eta(jj)$', y=1, size =25)
plt.xlabel(u'$\Delta \eta(jj)$', size =19)
plt.ylabel(u'Significance')
plt.xlim(3.75, 4.55)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.savefig('../../UndergraduateMonographDocument/SelectionCriteria/Figures/significance_deltaEta.pdf', format='pdf')
```

```python
dataMass = pd.read_csv('../massOptimization2/massOptimization.csv', skiprows=2)
```
