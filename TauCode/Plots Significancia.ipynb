```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-paper')
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
```

```python
data = pd.read_csv('../dEtaOptimization/dEtaOptimization.csv', skiprows=2)
```

```python
plt.plot(data['Value'], data['Significance'], 'o-')
plt.title(u'Significance vs $\Delta \eta(jj)$', y=1.08, size =25)
plt.xlabel(u'$\Delta \eta(jj)$', size =19)
plt.ylabel(u'Significance')
plt.xlim(37.5, 45.5)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
```

```python
dataMass = pd.read_csv('../massOptimization/massOptimization.csv', skiprows=2)
```

```python
plt.plot(dataMass['Value'], dataMass['Significance'], 'o-')
plt.title(u'Significance vs m(jj)', y=1.08, size =25)
plt.xlabel(u'm(jj)', size =19)
plt.ylabel(u'Significance')
plt.xlim(499.5, 1200.5)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
```
